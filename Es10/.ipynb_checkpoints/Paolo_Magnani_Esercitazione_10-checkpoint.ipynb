# Esercitazione 10

L'esercitazione 10 richiede di parallelizzare l'algoritmo genetico implementato nell'esercitazione 9, e di applicarlo al problema del _Commesso Viaggiatore_ tra le 50 città americane. In particolare, si richiede che i processi che lavorano in parallelo possano scambiarsi informazioni attraverso il protocollo MPI. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import seaborn as sns
color_palette = sns.color_palette("bright")
```

Dato che l'algoritmo genetico implementato era già parallelizzabile, l'unico cambio è stato quello di inizializzare il generatore di numeri casuali in funzione del numero di _rank_ con cui viene etichettato ciascun processo, così da ottenere sequenze pseudocasuali differenti. Dato che la mia macchina di lavoro è un pc portatile con 4 core ho potuto eseguire il codice su 4 processi distinti, etichettati con un numero di rank da $0$ a $3$.

## Processi non comunicanti
Per ciascuno dei 4 processi, ho generato una popolazione di $N=1000$ individui e l'ho fatta evolvere per $M=10^5$ generazioni. L'algoritmo converge rapidamente in ogni processo, salvo poi rimanere intrappolato in un minimo locale per molte generazioni ($\sim 100$). 3 processi su 4 convergono autonomamente verso lo stesso minimo.

```python
x = [[]for _ in range(4)]
y = [[]for _ in range(4)]
best = [[]for _ in range(4)]
gen = [[]for _  in range(4)]
mean = [[]for _ in range(4)]

for i in range(4):
    x[i], y[i] = np.loadtxt("risultati/indipendent/path_rank_"+str(i)+".dat", usecols=(1,2), unpack='true', delimiter=',')
    gen[i], best[i], mean[i] = np.loadtxt("risultati/indipendent/distance_rank_"+str(i)+".dat", usecols=(0,1,2), unpack='true', delimiter=',')
    x[i] = np.append(x[i],x[i][0])
    y[i] = np.append(y[i],y[i][0])

fig, ax = plt.subplots(2,2, figsize=(15,15))

for i in range(4):
    ax[i // 2, i % 2].plot(x[i], y[i], lw=1.5, marker='o')
    ax[i // 2, i % 2].set_title('Processo '+str(i)+' - Migliore distanza: '+str(best[i][-1]))
    ax[i // 2, i % 2].set_xlabel('Latitudine')
    ax[i // 2, i % 2].set_ylabel('Longitudine')
    ax[i // 2, i % 2].grid(True)
plt.show()

fig, ax = plt.subplots(2,2, figsize=(15,15))

fig.suptitle('Migliore distanza in funzione delle generazioni (<10000)', size=16)

for i in range(4):
    ax[i // 2, i % 2].plot(gen[i], best[i])
    ax[i // 2, i % 2].set_title('Processo '+str(i))
    ax[i // 2, i % 2].set_xlabel('# generazioni')
    ax[i // 2, i % 2].set_ylabel('$L^{(1)}$')
    ax[i // 2, i % 2].set_xlim(0, 10000)
    ax[i // 2, i % 2].set_ylim(250, 300)
    ax[i // 2, i % 2].grid(True)
plt.show()
```

## Processi comunicanti

Come nel caso dei processi non comunicanti, per ogni processo ho generato una popolazione di $N=1000$ individui e l'ho fatta evolvere per $M=10^5$ generazioni. Ogni $N_{migr}=1000$ ho fatto comunicare i 4 processi a coppie, scambiando le coppie alla comunicazione successiva. Durante la comunicazione, ciascun processo invia e riceve il miglior percorso al e dal processo con cui sta comunicando. La comunicazione avviene prima del _crossing over_, cosicché, se è stato ricevuto un percorso migliore dei propri, quest'ultimo ha già la possibilità di replicarsi nella popolazione.

```python
for i in range(4):
    x[i], y[i] = np.loadtxt("risultati/communicating/path_rank_"+str(i)+".dat", usecols=(1,2), unpack='true', delimiter=',')
    best[i] = np.loadtxt("risultati/communicating/distance_rank_"+str(i)+".dat", usecols=(1), unpack='true', delimiter=',')
    x[i] = np.append(x[i],x[i][0])
    y[i] = np.append(y[i],y[i][0])

fig, ax = plt.subplots(2,2, figsize=(15,15))

for i in range(4):
    ax[i // 2, i % 2].plot(x[i], y[i], lw=1.5, marker='o')
    ax[i // 2, i % 2].set_title('Processo '+str(i)+' - Migliore distanza: '+str(best[i][-1]))
    ax[i // 2, i % 2].set_xlabel('Latitudine')
    ax[i // 2, i % 2].set_ylabel('Longitudine')
    ax[i // 2, i % 2].grid(True)
plt.show()

fig, ax = plt.subplots(2,2, figsize=(15,15))

fig.suptitle('Migliore distanza in funzione delle generazioni (<10000)', size=16)

for i in range(4):
    ax[i // 2, i % 2].plot(gen[i], best[i])
    ax[i // 2, i % 2].set_title('Processo '+str(i))
    ax[i // 2, i % 2].set_xlabel('# generazioni')
    ax[i // 2, i % 2].set_ylabel('$L^{(1)}$')
    ax[i // 2, i % 2].set_xlim(0, 10000)
    ax[i // 2, i % 2].set_ylim(250, 300)
    ax[i // 2, i % 2].grid(True)
plt.show()
```

La comunicazione migliora la ricerca di un minimo, quantomeno nei primi scambi: dai grafici si può vedere che il processo di rank $0$ ha raggiunto un minimo più basso di tutti gli altri processi e lo scambia a $N_i = 1000$ con il processo $1$. Il processo $1$ conserva l'informazione e allo scambio per $N_i = 2000 $ tale informazione raggiunge, attraverso le comunicazioni a coppie scambiate, anche i processi $2$ e $3$. Tutti i processi si trovano così a covergere verso lo stesso minimo, più basso rispetto a quelli trovati in assenza di comunicazione.
