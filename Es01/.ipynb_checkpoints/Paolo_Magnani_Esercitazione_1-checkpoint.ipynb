## Esercizio 1.01
L'esercizio 1.01 richiede l'implementazione della media a blocchi.

### Punto 1

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
color_palette = sns.color_palette("bright")

x, sum41, error = np.loadtxt("1011.dat", usecols=(0,1,2), delimiter=',', unpack='true')
plt.errorbar(x,sum41,yerr=error)
plt.axhline(y=0, color="r")
plt.title('Valor medio di r in funzione del numero di lanci')
plt.xlabel('# lanci')
plt.ylabel(r'$<r>- \frac{1}{2}$')
plt.grid(True)
plt.show()
```

### Punto 2


```python
x2, sum42, error2 = np.loadtxt("1012.dat", usecols=(0,1,2), delimiter=',', unpack='true')
plt.errorbar(x2,sum42,yerr=error2)
plt.axhline(y=0, color="r")
plt.title('Varianza di r in funzione del numero di lanci')
plt.xlabel('# lanci')
plt.ylabel(r'$<(r-\frac{1}{2})^2>-\frac{1}{12}$')
plt.grid(True)
plt.show()
```

### Punto 3
Faccio un bell'istogramma del chi quadro.

```python
chi = np.loadtxt("1013.dat")
rep = np.arange(100)
media = np.mean(chi)

plt.plot(rep,chi, marker='o')
plt.title('Andamento del $\chi^2$')
plt.xlabel('# ripetizioni')
plt.ylabel('$\chi^2$')
plt.axhline(y=media, color="r", label="media: "+str(round(media, 1)))
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import chi2
media=100

x_chi = np.linspace(chi2.ppf(0.01, media), chi2.ppf(0.99, media), 100)
plt.plot(x_chi, chi2.pdf(x_chi, media), lw=4, c='r', alpha=0.6, label='distribuzione del $\chi^2$ di media 100')
plt.hist(chi, density=1, label='distribuzione dei dati')
plt.title('Confronto tra i dati ottenuti e la distribuzione del $\chi^2$')
plt.xlabel('Valori di $\chi^2$')
plt.ylabel('Occorrenze')
plt.grid(1)
plt.show()
```
