# Esercitazione 9

L'esercitazione 9 richiede di implementare un algoritmo genetico che risolva il problema del _Commesso viaggiatore_, quello cioè di trovare la più breve linea spezzata chiusa (il minor percorso) tra punti (città) posizionati su un piano. In particolare, si richiede che l'algoritmo trovi il percorso migliore tra 34 città posizionate randomicamente su una circonferenza e tra 34 città posizionate randomicamente in un quadrato.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import seaborn as sns
color_palette = sns.color_palette("bright")
```

Ho implementato un algoritmo che crea una "popolazione" di $N$ individui, detti _cromosomi_, contenenti ciascuno una sequenza casuale delle 34 città (ciascuna etichettata con un numero e senza ripetizioni). Ad ogni cromosoma è associata una funzione costo $L^{(1)}$, che misura la distanza euclidea tra tutte le città della sequenza contenuta nel cromosoma. L'algoritmo procede per $M$ cicli, definiti _generazioni_, in cui si ripetono le seguenti operazioni:
- la popolazione viene ordinata in ordine di funzione di costo crescente;
- attraverso una selezione truccata, definita da $N\times r^7$, con $r$ numero casuale estratto uniformemente tra $[0,1)$, si selezionano a coppie $N$ individui, che al $50\%$ producono attraverso _crossing over_ individui figli, che sostituiscono i genitori;
- la nuova popolazione così creata sostituisce la precedente, viene ricalcolata la funzione costo e viene riordinata;
- ciascun membro della popolazione subisce al $10\%$ delle mutazioni tra le $4$ implementate.

Il _crossing over_ consiste dello scambiare la parte finale della propria sequenza, di lunghezza selezionata in modo randomico, con un altro cromosoma, acquisendo al posto della coda perduta l'ordine delle città presente nell'altro cromosoma (evitando le ripetizioni).

Le quattro mutazioni consistono in:
- scambio di due città vicine nella sequenza;
- shift di $n$ città di $m$ posizioni, con $m$ e $n$ estratti uniformemente;
- permutazione di $k$ città vicine nella sequenza, con $k$ numero casuale da distribuzione uniforme;
- inversione di $j$ città vicine nella sequenza, con $j$ estratto uniformemente.

## Città posizionate su una circonferenza

Ho generato una popolazione di $N=1000$ percorsi e ho fatto evolvere il sistema per $M=100$ generazioni. In virtù della grande dimensione della popolazione, che ha garantito ergodicità, e della semplicità del problema (la soluzione è il cammino lungo la circonferenza) la convergenza verso la soluzione è stata rapida, trovando la soluzione migliore già alla 37sima generazione. Come è possibile vedere dai grafici, la funzione costo mediata sulla metà migliore della popolazione segue l'andamento della migliore distanza $L^{(1)}$ con un ritardo di poche ($\leq 3$) generazioni, segno che l'algoritmo trasmette molto rapidamente la sequenza migliore alla popolazione.

```python
x, y = np.loadtxt("risultati/best_path.dat", usecols=(1,2), delimiter=',', unpack='true')
gen, best, mean = np.loadtxt("risultati/best_distance.dat", usecols=(0,1,2), delimiter=',', unpack='true')
x = np.append(x, x[0])
y = np.append(y, y[0])

fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(15,5))

fig.suptitle("Città posizionate su una circonferenza", size=16)

ax0.plot(gen, best, color='g')
ax0.set_title("Migliore distanza $L^{(1)}$")
ax0.set_xlabel('# generazioni')
ax0.set_ylabel('$L^{(1)}$')
ax0.grid(True)

ax1.plot(gen, mean, color='C4')
ax1.set_title("Migliore distanza media")
ax1.set_xlabel('# generazioni')
ax1.set_ylabel(r"$\langle L^{(1)}\rangle$")
ax1.grid(True)

plt.show()

plt.plot(x, y, lw=1.5, marker='s')
plt.title("Migliore distanza: $L^{(1)} = $" + str(best[-1]) )
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

## Città posizionate in un quadrato

Ho generato una popolazione di $N=1000$ percorsi e ho fatto evolvere il sistema per $M=10^4$ generazioni, aumentando di un fattore $100$ le generazioni rispetto al caso delle città posizionate su un cerchio perché non sapevo la soluzione corretta (e il codice non era computazionalmente impegnativo dato che l'esecuzione non ha richiesto circa 1 minuto). In effetti, l'algoritmo converge molto rapidamente a distanze minori di $6$, ma tuttavia rimane intrappolato in minimi locali molto vicini al minimo globale, da cui esce solo dopo molte generazioni. L'algoritmo non è pertanto ottimizzato per la ricerca del minimo globale una volta che rimane per più generazioni in un minimo locale.

```python
x, y = np.loadtxt("risultati/best_path_square.dat", usecols=(1,2), delimiter=',', unpack='true')
gen, best, mean = np.loadtxt("risultati/best_distance_square.dat", usecols=(0,1,2), delimiter=',', unpack='true')
x = np.append(x, x[0])
y = np.append(y, y[0])

fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(15,5))

fig.suptitle("Città posizionate in un quadrato", size=16)

ax0.plot(gen, best, color='g')
ax0.set_title("Migliore distanza $L^{(1)}$")
ax0.set_xlabel('# generazioni')
ax0.set_ylabel('$L^{(1)}$')
ax0.grid(True)

ax1.plot(gen, mean, color='C4')
ax1.set_title("Migliore distanza media")
ax1.set_xlabel('# generazioni')
ax1.set_ylabel(r"$\langle L^{(1)}\rangle$")
ax1.grid(True)

plt.show()

plt.plot(x, y, lw=1.5, marker='o')
plt.title("Migliore distanza: $L^{(1)} = $" + str(best[-1]) )
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
