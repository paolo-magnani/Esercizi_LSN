```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import seaborn as sns
color_palette = sns.color_palette("bright")
```

## Citt√† posizionate su una circonferenza

```python
x, y = np.loadtxt("risultati/best_path.dat", usecols=(1,2), delimiter=',', unpack='true')
gen, best, mean = np.loadtxt("risultati/best_distance.dat", usecols=(0,1,2), delimiter=',', unpack='true')
x = np.append(x, x[0])
y = np.append(y, y[0])

fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(15,5))

fig.suptitle("Citt√† posizionate su una circonferenza", size=16)

ax0.plot(gen, best, color='g')
ax0.set_title("Migliore distanza $L^{(1)}$")
ax0.set_xlabel('# generazioni')
ax0.set_ylabel('$L^{(1)}$')
ax0.grid(True)

ax1.plot(gen, mean, color='C4')
ax1.set_title("Migliore distanza media")
ax1.set_xlabel('# generazioni')
ax1.set_ylabel(r"$\langle L^{(1)}\rangle$")
ax1.grid(True)

plt.show()

plt.plot(x, y, lw=1.5, marker='s')
plt.title("Migliore distanza: $L^{(1)} = $" + str(best[-1]) )
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
