## Esercizio 7.2
L'esercizio 7.2 richiede di studiare l'andamento dell'autocorrelazione tra i dati dell'energia potenziale per particella $U/N$ e dell'errore sugli stessi in funzione della dimensione dei blocchi con vengono suddivisi. Ho effettuato la simulazione attraverso l'algoritmo di Metropolis Monte Carlo per la dinamica molecolare implementato nel codice MD_MC. Per prima cosa, sono partito da una configurazione casuale delle particelle e ho equilibrato il sistema su $M=10^4$ step, registrando la configurazione finale delle molecole. Con il sistema equilibrato ho fatto evolvere il sistema per $5\times10^5$ step, registrando il valore di $U/N$ ad ogni punto senza suddividere i dati in blocchi. Ho effettuato la simulazione per le tre fasi dell'Argon:
1. fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$ (raggio di cutoff: $r_c = 2.2$)
2. fase liquida: $\rho^\star = 0.8$, $T^\star = 1.1$ (raggio di cutoff: $r_c = 2.5$)
3. fase gassosa: $\rho^\star = 0.05$, $T^\star = 1.2$ (raggio di cutoff: $r_c = 5.0$)

(dove i valori sono definiti in unit√† ridotte di Lennard-Jones).

```python
def autocorrelation(v, size):
    autoc = np.zeros(size)
    l = v.size - size
    mu1 = np.mean(v[:l])
    musq = np.mean(v[:l]**2)
    var = musq - mu1**2
    
    for i in range(size):
        mu2 = np.mean(v[i:l+i])
        mu12 = np.mean(v[:l]*v[i:l+i])
        
        autoc[i] = (mu12- (mu1*mu2))/var
        
    return autoc

#stima dell'errore
def error(sum, sum2, n):
    if(n==0): return 0
    else: return np.sqrt((sum2-(sum**2))/n)

def sigma(r,N):

    L = int(r.size/N)
    ave = 0
    av2 = 0

    for i in range(N):
        sum = 0

        for j in range(L):
            k = j+(i*L)
            sum += r[k]

        ave += sum/L
        av2 += (sum/L)**2

    ave /= N
    av2 /= N
    return error(ave, av2, N-1)

L = np.array([x for x in range(10,5001) if (500000%x ==0)]) #divisori di 500000
blocchi = 5*10**5 // L #integers
```

Attraverso le funzioni di autocorrelazione e di errore a blocchi definite qui sopra, ho stimato l'autocorrelazione e l'andamento dell'errore sui dati raccolti nelle tre simulazioni. Per il calcolo dell'autocorrelazione, ho preso i prini $1000$ dati. Per la stima dell'errore, ho valutato la $\sigma_{mean}$ sui dati per blocchi di dimensione maggiore di $10$ e inferiore a $5000$.

### Fase solida: autocorrelazione e stima dell'errore

```python
epot = np.loadtxt("risultati/Autocorrelazione/Solid/output_epot.dat", usecols=(1), unpack='true')

autoc = autocorrelation(epot, 1000)

y = []
for i in range(blocchi.size):
    y.append(sigma(epot, blocchi[i]))

fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(15,5))

fig.suptitle("Fase solida", size=14)

ax0.plot(autoc)
ax0.set_title("Autocorrelazione tra i dati di U/N")
ax0.set_xlabel("# dati per blocco")
ax0.set_ylabel("Autocorrelazione")
ax0.grid(True)

ax1.plot(L,y)
ax1.set_title("Andamento dell'errore")
ax1.set_xlabel('# dati per blocco')
ax1.set_ylabel("Stima dell'errore")
ax1.grid(True)

plt.show()
```

Dal grafico √® possibile vedere che l'autocorrelazione tra i primi $1000$ dati va a zero velocemente: sarebbe dunque sufficiente prendere blocchi di dimensione $\geq200$ per avere blocchi indipendenti su cui calcolare la media a blocchi. Il grafico sulla stima dell'errore mostra invece che per blocchi di dimensione $\leq 1000$ l'errore √® sottostimato, indicando la persistenza di effetti di correlazione tra i dati. Per le simulazioni successivo ho quindi scelto di prendere blocchi di dimensione $L=5000$.

### Fase liquida: autocorrelazione e stima dell'errore

```python
epot = np.loadtxt("risultati/Autocorrelazione/Liquid/output_epot.dat", usecols=(1), unpack='true')

autoc = autocorrelation(epot, 1000)

y = []
for i in range(blocchi.size):
    y.append(sigma(epot, blocchi[i]))

fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(15,5))

fig.suptitle("Fase liquida", size=14)

ax0.plot(autoc)
ax0.set_title("Autocorrelazione tra i dati di U/N")
ax0.set_xlabel("# dati per blocco")
ax0.set_ylabel("Autocorrelazione")
ax0.grid(True)

ax1.plot(L,y)
ax1.set_title("Andamento dell'errore")
ax1.set_xlabel('# dati per blocco')
ax1.set_ylabel("Stima dell'errore")
ax1.grid(True)

plt.show()
```

I grafici mostrano un andamento analogo a quello dei dati raccolti nella fase solida ma con una discesa al valore 0 dell'autocorrelazione e una saturazione dell'errore meno rapida rispetto alla fase solida. Anche per le simulazioni della fase liquida ho scelto una dimensione dei blocchi di $L=5000$.

### Fase gassosa: autocorrelazione e stima dell'errore

```python
epot = np.loadtxt("risultati/Autocorrelazione/Gas/output_epot.dat", usecols=(1), unpack='true')

autoc = autocorrelation(epot, 1000)

y = []
for i in range(blocchi.size):
    y.append(sigma(epot, blocchi[i]))

fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(15,5))

fig.suptitle("Fase gassosa", size=14)

ax0.plot(autoc)
ax0.set_title("Autocorrelazione tra i dati di U/N")
ax0.set_xlabel("# dati per blocco")
ax0.set_ylabel("Autocorrelazione")
ax0.grid(True)

ax1.plot(L,y)
ax1.set_title("Andamento dell'errore")
ax1.set_xlabel('# dati per blocco')
ax1.set_ylabel("Stima dell'errore")
ax1.grid(True)

plt.show()
```

I dati della fase gassosa hanno un'autocorrelazione che decresce molto rapidamente, cos√¨ come molto rapidamente satura l'errore. Come negli altri casi, la dimensione dei blocchi di simulazione successivi √® stata pari a $L=5000$.

## Esercizio 7.4
Nell'esercizio si richiede che si calcolino i valori medi di energia potenziale per particella $U/N$ e di pressione $P$ nelle tre diverse fasi dell'Argon presentate sopra attraverso l'utilizzo dell'algoritmo di Metropolis NVT. Si richiede anche di stimare, sia con il Molecular Dynamics NVE che con il Monte Carlo NVT, la funzione di distribuzione radiale $g(r)$ nelle tre fasi termodinamiche. In entrambi i casi ho equilibrato il sistema per $10^5$ steps prima di far ripartire la simulazione. Nel caso dell'algoritmo MC ho scelto un passo del Metropolis che desse una probabilit√† di accettazione della mossa al $50\%$. Nel caso della fase gassosa, in cui l'accettanza non scendeva al di sotto del $60\%$, ho scelto un passo pari a quello del arggio di cut-off.

### Fase solida

```python
x, pres, error = np.loadtxt("risultati/NVT/Solid/output_pres.dat", usecols=(0,2,3), unpack='true')
x2, epot, error2 = np.loadtxt("risultati/NVT/Solid/output_epot.dat", usecols=(0,2,3), unpack='true')

fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(15,5))

fig.suptitle("Fase solida: energia potenziale e pressione", size=14)

ax0.errorbar(x2,epot,yerr=error2)
ax0.set_xlabel('# blocchi')
ax0.set_ylabel("Energia potenziale per particella [r.u.]")
ax0.grid(True)

ax1.errorbar(x,pres,yerr=error, color='C4')
ax1.set_xlabel('# blocchi')
ax1.set_ylabel("Pressione [r.u.]")
ax1.grid(True)
#ax1.legend()

plt.show()
```

```python
x, sum, err = np.loadtxt("risultati/NVT/Solid/output_gfinal.dat", usecols=(0,1,2), unpack='true')
x2, sum2, err2 = np.loadtxt("risultati/NVE/Solid/output_gfinal.dat", usecols=(0,1,2), unpack='true')

plt.errorbar(x2, sum2, yerr=err2, label="Molecular Dynamics NVE")
plt.errorbar(x, sum, yerr=err, label="Monte Carlo NVT")
plt.title("Fase solida: distribuzione radiale $g(r)$")
plt.xlabel("raggio $r$ [r.u.]")
plt.ylabel("$g(r)$ [a.u.]")
plt.grid(True)
plt.legend()
plt.show()
```

L'algoritmo MC e l'algoritmo MD riproducono lo stesso andamento della funzione $g(r)$, che presenta picchi che lasciano intendere la struttura cristallina della fase solida.
