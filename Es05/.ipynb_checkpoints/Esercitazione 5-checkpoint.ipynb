## $\Psi_{100}$ and $\Psi_{210}$, uniform distribution, mean radius values

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
color_palette = sns.color_palette("bright")

x, mean, error = np.loadtxt("psi100_mean.dat", usecols=(0,1,2), delimiter=',', unpack='true')
plt.errorbar(x,mean,yerr=error)
plt.axhline(y=1.5, c='r', label='valore atteso')
plt.title('Raggio medio per lo stato $1s$')
plt.xlabel('# lanci')
plt.ylabel('<r>, $\Psi_{100}$ [Bohr radius units]')
plt.legend()
plt.grid(True)
plt.show()

x, mean, error = np.loadtxt("psi210_mean.dat", usecols=(0,1,2), delimiter=',', unpack='true')
plt.errorbar(x,mean,yerr=error)
plt.axhline(y=5, c='r', label='valore atteso')
plt.title('Raggio medio per lo stato $2p$')
plt.xlabel('# lanci')
plt.ylabel('<r>, $\Psi_{210}$ [Bohr radius units]')
plt.legend()
plt.grid(True)
plt.show()
```

## $\Psi_{100}$ and $\Psi_{210}$, uniform distribution, data 

```python
X, Y, Z = np.loadtxt("psi100_data.dat", usecols=(0,1,2), delimiter=',', unpack='true')

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(5, 45)

plt.show()
```
