# Esercitazione 5

L'esercitazione 5 richiede che venga stimato il raggio medio degli stati $\Psi_{100}$ e $\Psi_{210}$ per l'atomo di idrogeno attraverso l'implementazione dell'algoritmo di Metropolis. In particolare, si richiede che l'algoritmo di Metropolis campioni le distribuzioni di probabilit√† $|\Psi_{1,0,0}(x,y,z)|^2$ e $|\Psi_{2,1,0}(x,y,z)|^2$ attraverso una funzione di transizione $T(\vec{x}|\vec{y})$ uniforme e una funzione di transizione gaussiana.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
color_palette = sns.color_palette("bright")
```

### $\Psi_{100}$ e $\Psi_{210}$, distribuzione uniforme

```python
x, mean, error = np.loadtxt("risultati/psi100_mean.dat", usecols=(0, 1, 2), delimiter=',', unpack=True)
X, Y, Z = np.loadtxt("risultati/psi100_data.dat", usecols=(0, 1, 2), delimiter=',', unpack=True)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle("Stato $1s$", size=14)

axes[0].errorbar(x, mean, yerr=error)
axes[0].axhline(y=1.5, c='r', label='valore atteso')
axes[0].set_title('Raggio medio per lo stato $1s$')
axes[0].set_xlabel('# estrazioni')
axes[0].set_ylabel('<r>, $\Psi_{100}$ [Bohr radius units]')
axes[0].legend()
axes[0].grid(True)

axes[1].axis('off')
axes[1] = fig.add_subplot(122, projection='3d')
axes[1].scatter(X, Y, Z, c=Z, marker='.')
axes[1].set_title("Distribuzione spaziale dei punti estratti")
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_zlabel('z')
axes[1].view_init(5, 45)

plt.show()


x, mean, error = np.loadtxt("risultati/psi210_mean.dat", usecols=(0, 1, 2), delimiter=',', unpack=True)
X, Y, Z = np.loadtxt("risultati/psi210_data.dat", usecols=(0, 1, 2), delimiter=',', unpack=True)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle("Stato $2p$", size=14)

axes[0].errorbar(x, mean, yerr=error)
axes[0].axhline(y=5, c='r', label='valore atteso')
axes[0].set_title('Raggio medio per lo stato $2p$')
axes[0].set_xlabel('# estrazioni')
axes[0].set_ylabel('<r>, $\Psi_{210}$ [Bohr radius units]')
axes[0].legend()
axes[0].grid(True)

axes[1].axis('off')
axes[1] = fig.add_subplot(122, projection='3d')
axes[1].scatter(X, Y, Z, c=Z, marker='.')
axes[1].set_title("Distribuzione spaziale dei punti estratti")
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_zlabel('z')
axes[1].view_init(5, 45)

plt.show()
```

## Confronto tra le distribuzioni

```python
x, mean, error_u = np.loadtxt("risultati/psi100_mean.dat", usecols=(0,1,2), delimiter=',', unpack='true')
plt.errorbar(x,mean,yerr=error_u, label='distribuzione uniforme')
plt.axhline(y=1.5, c='C4', label='valore atteso')
x, mean, error_g = np.loadtxt("risultati/psi100_meanGauss.dat", usecols=(0,1,2), delimiter=',', unpack='true')
plt.errorbar(x,mean,yerr=error_g, label='distribuzione Gaussiana')
plt.title('Confronto per lo stato $1s$')
plt.xlabel('# lanci')
plt.ylabel('<r>, $\Psi_{100}$ [Bohr radius units]')
plt.legend()
plt.grid(True)
plt.show()

plt.plot(x,error_u, label='distribuzione uniforme')
plt.plot(x,error_g, label='distribuzione Gaussiana')
plt.title("Andamento dell'errore per lo stato $1s$")
plt.xlabel('# lanci')
plt.ylabel('errore [Bohr radius units]')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x, mean, error_u = np.loadtxt("risultati/psi210_mean.dat", usecols=(0,1,2), delimiter=',', unpack='true')
plt.errorbar(x,mean,yerr=error_u, label='distribuzione uniforme')
plt.axhline(y=5, c='C4', label='valore atteso')
x, mean, error_g = np.loadtxt("risultati/psi210_meanGauss.dat", usecols=(0,1,2), delimiter=',', unpack='true')
plt.errorbar(x,mean,yerr=error_g, label='distribuzione Gaussiana')
plt.title('Confronto per lo stato $2p$')
plt.xlabel('# lanci')
plt.ylabel('<r>, $\Psi_{100}$ [Bohr radius units]')
plt.legend()
plt.grid(True)
plt.show()

plt.plot(x,error_u, label='distribuzione uniforme')
plt.plot(x,error_g, label='distribuzione Gaussiana')
plt.title("Andamento dell'errore per lo stato $2p$")
plt.xlabel('# lanci')
plt.ylabel('errore [Bohr radius units]')
plt.legend()
plt.grid(True)
plt.show()
```
