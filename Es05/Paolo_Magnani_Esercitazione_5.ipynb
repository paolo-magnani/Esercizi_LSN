# Esercitazione 5

L'esercitazione 5 richiede che venga stimato il raggio medio degli stati $\Psi_{100}$ e $\Psi_{210}$ per l'atomo di idrogeno attraverso l'implementazione dell'algoritmo di Metropolis. In particolare, si richiede che l'algoritmo di Metropolis campioni le distribuzioni di probabilità $|\Psi_{1,0,0}(x,y,z)|^2$ e $|\Psi_{2,1,0}(x,y,z)|^2$ attraverso una funzione di transizione $T(\vec{x}|\vec{y})$ uniforme e una funzione di transizione gaussiana.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
color_palette = sns.color_palette("bright")
```

### $\Psi_{100}$ e $\Psi_{210}$, distribuzione uniforme
Ho stimato il valor medio $\langle r \rangle _{100} = \cfrac{3}{2}a_0 $  e $\langle r \rangle _{210} = 5a_0 $ (con $a_0=1$) campionando $M=10^6$ punti attraverso Metropolis: l'algoritmo parte dal punto $(0,0,0)$ e propone una mossa di passo $\delta$ per un numero estratto uniformemente tra $[-1,1]$, lungo i tre assi cartesiani. La mossa è automaticamente accettata se il punto proposto è a probabilità maggiore del punto di partenza, mentre nel caso opposto è accettata con probabilità pari al rapporto tra le densità di probabilità nei due punti. La distribuzione dei punti estratti segue così la densità di probabilità dell'orbitale. Il valore di $\delta$ è stato scelto per ottenere una probabilità di accettanza della mossa attorno al $50\%$.  Valor medio e relativo errore sono stati stimati con il metodo della media a blocchi con $N_{blocchi}=100$.

```python
x, mean, error = np.loadtxt("risultati/psi100_mean.dat", usecols=(0, 1, 2), delimiter=',', unpack=True)
X, Y, Z = np.loadtxt("risultati/psi100_data.dat", usecols=(0, 1, 2), delimiter=',', unpack=True)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle("Stato $1s$", size=14)

axes[0].errorbar(x, mean, yerr=error)
axes[0].axhline(y=1.5, c='r', label='valore atteso')
axes[0].set_title('Raggio medio per lo stato $1s$')
axes[0].set_xlabel('# estrazioni')
axes[0].set_ylabel('<r>, $\Psi_{100}$ [Bohr radius units]')
axes[0].legend()
axes[0].grid(True)

axes[1].axis('off')
axes[1] = fig.add_subplot(122, projection='3d')
axes[1].scatter(X, Y, Z, c=Z, marker='.')
axes[1].set_title("Distribuzione spaziale dei punti estratti")
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_zlabel('z')
axes[1].view_init(5, 45)

plt.show()


x, mean, error = np.loadtxt("risultati/psi210_mean.dat", usecols=(0, 1, 2), delimiter=',', unpack=True)
X, Y, Z = np.loadtxt("risultati/psi210_data.dat", usecols=(0, 1, 2), delimiter=',', unpack=True)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle("Stato $2p$", size=14)

axes[0].errorbar(x, mean, yerr=error)
axes[0].axhline(y=5, c='r', label='valore atteso')
axes[0].set_title('Raggio medio per lo stato $2p$')
axes[0].set_xlabel('# estrazioni')
axes[0].set_ylabel('<r>, $\Psi_{210}$ [Bohr radius units]')
axes[0].legend()
axes[0].grid(True)

axes[1].axis('off')
axes[1] = fig.add_subplot(122, projection='3d')
axes[1].scatter(X, Y, Z, c=Z, marker='.')
axes[1].set_title("Distribuzione spaziale dei punti estratti")
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_zlabel('z')
axes[1].view_init(5, 45)

plt.show()
```

Le stime della media e del relativo errore sono state calcolate attraverso il metodo della media a blocchi ($N_{blocchi}=100$). 

```python
x, mean, error = np.loadtxt("risultati/psi100_meanGauss.dat", usecols=(0, 1, 2), delimiter=',', unpack=True)
X, Y, Z = np.loadtxt("risultati/psi100_dataGauss.dat", usecols=(0, 1, 2), delimiter=',', unpack=True)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle("Stato $1s$", size=16)

axes[0].errorbar(x, mean, yerr=error)
axes[0].axhline(y=1.5, c='r', label='valore atteso')
axes[0].set_title('Raggio medio per lo stato $1s$')
axes[0].set_xlabel('# estrazioni')
axes[0].set_ylabel('<r>, $\Psi_{100}$ [Bohr radius units]')
axes[0].legend()
axes[0].grid(True)

axes[1].axis('off')
axes[1] = fig.add_subplot(122, projection='3d')
axes[1].scatter(X, Y, Z, c=Z, marker='.')
axes[1].set_title("Distribuzione spaziale dei punti estratti")
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_zlabel('z')
axes[1].view_init(5, 45)

plt.show()


x, mean, error = np.loadtxt("risultati/psi210_meanGauss.dat", usecols=(0, 1, 2), delimiter=',', unpack=True)
X, Y, Z = np.loadtxt("risultati/psi210_dataGauss.dat", usecols=(0, 1, 2), delimiter=',', unpack=True)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle("Stato $2p$", size=16)

axes[0].errorbar(x, mean, yerr=error)
axes[0].axhline(y=5, c='r', label='valore atteso')
axes[0].set_title('Raggio medio per lo stato $2p$')
axes[0].set_xlabel('# estrazioni')
axes[0].set_ylabel('<r>, $\Psi_{210}$ [Bohr radius units]')
axes[0].legend()
axes[0].grid(True)

axes[1].axis('off')
axes[1] = fig.add_subplot(122, projection='3d')
axes[1].scatter(X, Y, Z, c=Z, marker='.')
axes[1].set_title("Distribuzione spaziale dei punti estratti")
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_zlabel('z')
axes[1].view_init(5, 45)

plt.show()
```

## Confronto tra le distribuzioni

L'algoritmo di Metropolis campiona correttamente il valor medio delle distribuzioni sia con la funzione di transizione uniforme, sia con la funzione di transizione Gaussiana. Per lo stato $1s$, la cui $|\Psi|^2$ è una gaussiana, gli errori diminuiscono usando una $T(\vec{x}|\vec{y})$ del secondo tipo (nello stesso modo dell'_importance sampling_).

```python
x, mean_u, error_u = np.loadtxt("risultati/psi100_mean.dat", usecols=(0,1,2), delimiter=',', unpack='true')
x2, mean_g, error_g = np.loadtxt("risultati/psi100_meanGauss.dat", usecols=(0,1,2), delimiter=',', unpack='true')

fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(15,5))

fig.suptitle("Stato $1s$", size=16)

ax0.errorbar(x,mean_u,yerr=error_u, label='distribuzione uniforme')
ax0.errorbar(x2,mean_g,yerr=error_g,  label='distribuzione Gaussiana')
ax0.axhline(y=1.5, color="C4", label='valore atteso', lw=2)
ax0.set_xlabel('# estrazioni')
ax0.set_ylabel("<r>, $\Psi_{100}$ [Bohr radius units]")
ax0.grid(True)
ax0.legend()

ax1.plot(x, error_u, marker='x', linestyle='none', label='distribuzione uniforme')
ax1.plot(x2, error_g, marker='.',linestyle='none',  label='distribuzione Gaussiana')
ax1.set_xlabel('# estrazioni')
ax1.set_ylabel("Stima dell'errore [Bohr radius units]")
ax1.grid(True)
ax1.legend()

plt.show()
```

Per lo stato $2p$, la cui densità di probabilità non è più gaussiana, la differenza tra la stima con distribuzione gaussiana non è più apprezzabile.

```python
x, mean_u, error_u = np.loadtxt("risultati/psi210_mean.dat", usecols=(0,1,2), delimiter=',', unpack='true')
x2, mean_g, error_g = np.loadtxt("risultati/psi210_meanGauss.dat", usecols=(0,1,2), delimiter=',', unpack='true')

fig, [ax0, ax1] = plt.subplots(1, 2, figsize=(15,5))

fig.suptitle("Stato $2p$",size=16)

ax0.errorbar(x,mean_u,yerr=error_u, label='distribuzione uniforme')
ax0.errorbar(x2,mean_g,yerr=error_g,  label='distribuzione Gaussiana')
ax0.axhline(y=5, color="C4", label='valore atteso', lw=2)
ax0.set_xlabel('# estrazioni')
ax0.set_ylabel("<r>, $\Psi_{100}$ [Bohr radius units]")
ax0.grid(True)
ax0.legend()

ax1.plot(x, error_u, marker='x', linestyle='none', label='distribuzione uniforme')
ax1.plot(x2, error_g, marker='.',linestyle='none',  label='distribuzione Gaussiana')
ax1.set_xlabel('# estrazioni')
ax1.set_ylabel("Stima dell'errore [Bohr radius units]")
ax1.grid(True)
ax1.legend()

plt.show()
```

## Coordinate lontane dall'origine
